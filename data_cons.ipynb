```python
apart = pd.read_csv('data/apartments.csv')
apart.head()
```

```python
apart['rate'].value_counts()
```

```python
apart.info()
```

```python
apart['price'] = apart['price'].str.replace(' ', '')
```

```python
apart['price'] = pd.to_numeric(apart['price'], errors='coerce')
apart['price'] = apart['price'].astype('Int64')
apart.info()
```

```python
apart['location'].value_counts()
```

```python
Location = []
City = []



location = apart['location']
location

for loc in location:
    try:
        house_loc = loc.split(',')
        City.append(house_loc[-1])
        Location.append(house_loc[-2])
    except:
        City.append(None)
        Location.append(None)
```

```python
apart['City'] = City
apart['Location'] = Location
apart.head()
```

```python
apart.drop(apart[['Unnamed: 0','title', 'location']], axis=1, inplace=True)
apart.head()
```

```python
mask = apart['rate'] == 'Per Day'

apart.loc[mask, 'price'] *= 30

apart.loc[mask, 'rate'] = 'Per Month'
```

```python
apart['rate'].value_counts()
```

```python
apart.drop(columns='rate',inplace=True)
```

```python
new_col_name = {'bedrooms': 'Bedrooms', 'bathrooms': 'Bathrooms', 'price': 'Price'}
apart.rename(columns=new_col_name, inplace=True)
```

```python
apart.head()
```

```python
apart.to_csv('data/apartment_final.csv')
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
house_url = 'https://kenyapropertycentre.com/for-rent/houses/townhouses/machakos/syokimau-mulolongo/26305-executive-4-bedroom-house-all-bedroom-en-suite-dsq'
house_response = requests.get(house_url)
if house_response.status_code != 200:
    print("Request failed")
```

```python
house_soup = BeautifulSoup(house_response.text, 'lxml')
```

```python
price = house_soup.find('div', class_='wp-block-content clearfix hidden-md hidden-sm hidden-lg voffset-15').find('span', class_='pull-left').find_all('span', class_='price')
price = price[1].text.strip()
price
```

```python
loc = house_soup.find('div', class_='col-sm-8 f15 property-details').find('address').text.strip().split(',')

City = loc[2]
Location = loc[1]
print(Location, City)
```

```python
interior = house_soup.find('div', class_='wp-block-footer style2 mt-15').find('ul', class_='aux-info').find_all('li')
interior
```

```python
bedrooms = interior[0].find('span').text.strip()
bedrooms
```

```python
bathrooms = interior[1].find('span').text.strip()
bathrooms
```

```python
toilets = interior[2].find('span').text.strip()
toilets
```
